#Logical Operator

```python
#Logical operator
x=2
y=3
if(x>5 and x<10):
  print(True)
else:
  print(False)

if(x>0 or x<1):
  print(True)
else:
  print(False)

if not(x>0 or x<1): #Reverse the result
  print(True)
else:
  print(False)

if x is x:          #Address Checking of the variable insted of value
  print(True)
if x is not y:
  print(False)

p=['hi','bye']
q=["hi","bye"]
if p[0] in q:          #If the value in p is present in q
  print(True)
if p[0] and p[1] in q:
  print(True)
else:
  print(False)
```

#Tuples

```python
#A tuple is a collection which is ordered and unchangeable.
thistuple = ("apple", "banana", "cherry")
print(thistuple)
print(thistuple[0])
thistuple = ("apple", "banana", "cherry", "apple", "cherry")
print(thistuple)
print(len(thistuple))
tuple1 = ("abc", 34, True, 40, "male")
print(tuple1)
print(thistuple[2:4])

#Changing the value of tuple
x = ("apple", "banana", "cherry")
y = list(x)
y.append("kiwi")
x = tuple(y)
print(x)
```
